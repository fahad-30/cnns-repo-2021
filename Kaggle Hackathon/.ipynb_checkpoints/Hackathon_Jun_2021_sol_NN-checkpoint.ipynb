```python
import torch
torch.cuda.is_available()
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
# device = torch.device("cpu")
device
```

```python
import pandas as pd
train = pd.read_csv('tabular-playground-series-jun-2021/train.csv')
```

```python
train.head(10)
```

```python
from torch import nn,optim
import torch.nn.functional as F
from torch.autograd import Variable

class classifier(nn.Module):
    def __init__(self):
        super().__init__()
        self.fc1 = nn.Linear(75,32)
        self.fc2 = nn.Linear(32,9)
    
    def forward(self,x):
        x.to(device)
        x.double()
        x = F.relu(self.fc1(x))
        x = F.log_softmax(self.fc2(x),dim = 1)
        
        return x
```

```python
train_x = train.drop('target',axis = 'columns')
train_x.drop('id',axis='columns',inplace = True)
for col in train_x:
    train_x[col] /= train_x[col].max()
# train_x
train_x = Variable(torch.tensor(train_x.values)).to(device)
train_x = train_x.type(torch.DoubleTensor)


one_hot_y = pd.get_dummies(train['target'])
one_hot_y =  Variable(torch.tensor(one_hot_y.values)).to(device)
train_y = torch.tensor(one_hot_y.max(1)[1])
train_y = train_y.view(200000,1)
train_x.to(device)
```

```python
lr = 0.001
epochs = 10

model = classifier()

model.to(device)
crit = nn.NLLLoss()
opt = optim.SGD(model.parameters(),lr = lr)
```

```python
model.double()
train_x.to(device)
train_y.to(device)
print(train_x[0])

out = model(train_x[0].view(1,75).to(device))
print(out)
out = out.to(device)
# loss = crit(out.view(1,9),train_y[1].view(1,9))
loss = crit(out,train_y[0])
loss
```

```python
import os
if os.path.exists('NN_checkpoint.pth'):
    print("Path Exists")
    state_dict = torch.load('NN_checkpoint.pth')
    model.load_state_dict(state_dict)
    pass
else:

    print_every = 10000 #inputs 10,000

    for e in range(epochs):
        running_loss = 0
        steps=0
        flag = False
        model.train()
        for i in range(train_y.shape[0]):
            steps+=1
            opt.zero_grad()

            out = model.forward(train_x[i].view(1,75).to(device))
            loss = crit(out,train_y[i])
            loss.backward()
            opt.step()

            running_loss += loss.item()


            if steps%print_every == 0:
                print("Epoch: {}/{}.. ".format(e+1, epochs),
                      "Step:{}/{}.. ".format(steps,train_y.shape[0]),
                          "Training Loss: {:.3f}.. ".format((running_loss/print_every)))
                running_loss = 0
    torch.save(model.state_dict(), 'NN_checkpoint.pth')

```

```python
accuracy = 0
test_loss = 0
model.eval()
for i in range(train_y.shape[0]):
    with torch.no_grad():
        out = model.forward(train_x[i].view(1,75).to(device))
        test_loss += crit(out,train_y[i]).item()
        exp_out = torch.exp(out)
        equality = (train_y[i].data == exp_out.max(1)[1])
#         if i%print_every == 0:
#             print(equality.type_as(torch.FloatTensor()))
        accuracy += equality.type_as(torch.FloatTensor()).item()
        

```

```python
print("Test Loss: {:.3f}.. ".format(test_loss/(train_y.shape[0])),
      "Test Accuracy: {:.3f}".format(accuracy/(train_y.shape[0])))
```
